## Najwa Shihab Youtube Channel Market Share

Here we tried to analyse how many shares Najwa Shihab Youtube Channel had gained compared to other channels on related Category. Based on socialblade (https://socialblade.com/youtube/channel/UCo8h2TY_uBkAVUIc14m_KCA) this channel is in News & Politics Category. 

#### Methodology

The methodology we used was pretty simple, since socialblade didn't explicitly provide Top News & Politics Channel in Indonesia so we scraped the data from Top 250 Youtube Channel in Indonesia on socialblade (https://socialblade.com/youtube/top/country/id) to get the most popular Youtube Channel based on its views and subscribers. This Top 250 are consists of many channel ranging from different categories, here we want to focus on News & Entertainment Category since Najwa Shihab Channel belong to that category. After

After that we scraped another list, Top 250 News & Politics Channel Worldwide (https://socialblade.com/youtube/top/category/news). After we got the data, we inner joined the data based on similar value which is the channel name. Finally we managed to get the Top News & Politics Channel in Indonesia

```python
import pandas as pd 
import numpy as np
from pandas.plotting import table
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```


##### Data as per July, 12th 2019

```python
df_top_indo = pd.read_csv('250 Top Indonesia.csv')
df_news = pd.read_csv('250 Top News Channel.csv')
```

##### Top 250 Youtube Channel in Indonesia 

```python
df_top_indo.head()
```

##### Top 250 News & Politics Channel Worldwide 

```python
df_news.head()
```

We inner joined the data and managed to get 13 News and Politics Channel in Indonesia. Using this data we can calculate how big is Najwa Shihab Youtube Channel share is

```python
df_merge = pd.merge(left=df_news, right=df_top_indo, left_on='name', right_on='name')
df_merge
```

```python
df_merge.dtypes
```

```python
df_merge.drop(['rade_y', 'uploads_y', 'subs_y', 'views_y'], axis=1, inplace=True)
```

```python
df_merge['uploads_x'] = df_merge['uploads_x'].str.replace(',', '')
df_merge['subs_x'] = df_merge['subs_x'].str.replace(',', '')
df_merge['views_x'] = df_merge['views_x'].str.replace(',', '')
```

```python
df_merge['uploads_x'] = df_merge['uploads_x'].astype(int)
df_merge['subs_x'] = df_merge['subs_x'].astype(int)
df_merge['views_x'] = df_merge['views_x'].astype(int)
```

```python
df_merge.dtypes
```

We calculated the data to get Total amount of subs, uploads, and views

```python
sums = df_merge.select_dtypes(pd.np.number).sum().rename('Total')
df_merge.append(sums)
```

#### Uploads

Number of videos uploaded on News and Politics Channel in Indonesia. Berita Satu has uploaded more than 150,000 videos and Sidul Tube has the least uploaded videos by far.

```python
plt.figure(figsize=(10,6))
ax_2 = sns.barplot(x = 'uploads_x', y = 'name', data=df_merge.sort_values(by=['uploads_x'], ascending=False))

ax_2.set_title('Number of uploaded videos')

rects_2 = ax_2.patches

for rect in rects_2:
    y_value = rect.get_y() + rect.get_height() / 2
    x_value = rect.get_width()
    
    space = 5
    
    va = 'top'
    
    label = '{:.0f}'.format(x_value)
    
    plt.annotate(
        label, 
        (x_value, y_value),
        xytext=(10, space),
        textcoords='offset points',
        ha='left',
        va = va)
```

### Based on Subs

##### Market Share

Here we try to analyse the market share by calculating the percentage of subscribers, as we seen that the highest subs is Daftar Popular with 21.4% subscriptions share. Najwa Shihab has 8.5% subscriptions share with more than 2.5 million subs.

```python
df_merge['perc_subs'] = (df_merge['subs_x'] / df_merge['subs_x'].sum())*100
df_merge.sort_values(by=['perc_subs'], ascending=False)
```

To gain clearer figure of how big is our shares, we can see the pie chart below that show each slice of the channel shares.

```python
fig = plt.figure(figsize=(20,12))

explode = (0, 0, 0, 0, 0, 0, 0, 0.12, 0, 0, 0, 0, 0, 0)  

# plot chart
ax_pie_subs = plt.subplot(121, aspect='equal')
df_merge.plot(kind='pie', y = 'perc_subs', ax=ax_pie_subs, autopct='%1.1f%%', pctdistance=0.8, explode=explode,
        startangle=60, shadow=False, labels=df_merge['name'], legend = False, fontsize=9)

ax_pie_subs.set_title('Market Share based on Subscriptions')

fig.patch.set_facecolor('white')

# plot table
#ax_table_subs = plt.subplot(122)
plt.axis('off')
#tbl = table(ax_table_subs, df_merge, loc='center')
#tbl.auto_set_font_size(False)
#tbl.set_fontsize(14)
plt.show()
```

Another visualisations from the table above

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(x = 'subs_x', y = 'name', data=df_merge.sort_values(by=['subs_x'], ascending=False))

ax.set_title('Most subs on youtube channel, News Category')

rects = ax.patches

for rect in rects:
    y_value = rect.get_y() + rect.get_height() / 2
    x_value = rect.get_width()
    
    space = 5
    
    va = 'top'
    
    label = '{:.0f}'.format(x_value)
    
    plt.annotate(
        label, 
        (x_value, y_value),
        xytext=(10, space),
        textcoords='offset points',
        ha='left',
        va = va)
```

#### Based on views

##### Market Share

Here we try to analyse the market share by calculating the percentage of total video views, as we seen that the highest video views is CNN Indonesia with 14.7% video views share. Najwa Shihab has 4.3% video views share with more than 400 video views.

```python
df_merge['perc_views'] = (df_merge['views_x'] / df_merge['views_x'].sum())*100
df_merge.sort_values(by=['perc_views'], ascending=False)
```

To gain better understanding of how big is our shares, we can see the pie chart below that show each slice of the channel shares.

```python
fig = plt.figure(figsize=(20,12))

explode = (0, 0, 0, 0, 0, 0, 0, 0.15, 0, 0, 0, 0, 0, 0)  

# plot chart
ax_pie_subs = plt.subplot(121, aspect='equal')
df_merge.plot(kind='pie', y = 'perc_views', ax=ax_pie_subs, autopct='%1.1f%%', pctdistance=0.8, explode=explode,
        startangle=60, shadow=False, labels=df_merge['name'], legend = False, fontsize=9)

ax_pie_subs.set_title('Market Share based on Views')
fig.patch.set_facecolor('white')

# plot table
#ax_table_subs = plt.subplot(122)
plt.axis('off')
#tbl = table(ax_table_subs, df_merge, loc='center')
#tbl.auto_set_font_size(False)
#tbl.set_fontsize(14)
plt.show()
```
