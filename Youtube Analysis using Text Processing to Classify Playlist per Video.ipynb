```python
all_result
```

```python
JustPlaylist = all_result[['Playlist','viewCount']]


#df.groupby('DATE', as_index=False).sum()
```

```python
JustPlaylistSumALL = JustPlaylist.groupby('Playlist', as_index=False).agg({'viewCount': 'sum'})
```

```python
NumberOfVideos = JustPlaylist.groupby('Playlist', as_index = False).count()
```

```python
NumberOfVideos = NumberOfVideos.rename(columns={'viewCount': 'Total Video'})
```

```python
NumberOfVideos
```

```python
MeanOfViews = JustPlaylist.groupby('Playlist', as_index = False)['viewCount'].mean()
```

```python
MeanOfViews = MeanOfViews.rename(columns={'viewCount': 'Views Mean'})
```

```python
MeanOfViews['Views Mean'] = MeanOfViews['Views Mean'].round(2)
```

```python
MeanOfViews
```

```python
dfALL = pd.concat([JustPlaylistSumALL, NumberOfVideos['Total Video'], MeanOfViews['Views Mean']], axis=1, sort=False)
```

```python
dfALL
```

```python
PlaylistPijaru_sorted = JustPlaylistSumALL.sort_values(by=['viewCount'], ascending = False)
```
