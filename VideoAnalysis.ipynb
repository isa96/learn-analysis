```python
import cv2
import face_recognition
```

```python
input_video = cv2.VideoCapture('MataNajwa.mp4')
```

```python
length = int(input_video.get(cv2.CAP_PROP_FRAME_COUNT))
```

```python
#fourcc = cv2.VideoWriter_fourcc('M','P','E','G')
fourcc = cv2.VideoWriter_fourcc(*'XVID')
```

```python
output_video = cv2.VideoWriter('output3.avi', fourcc, 24, (1280, 720))
```

```python
NajwaShihab = face_recognition.load_image_file("najwashihab.jpg")
NajwaShihabEncoding = face_recognition.face_encodings(NajwaShihab)[0]
```

```python
Maruarar = face_recognition.load_image_file("maruar_sirait.jpg")
MaruararEncoding = face_recognition.face_encodings(Maruarar)[0]
```

```python
AkmalMarhali = face_recognition.load_image_file("akmalmarhali.jpg")
AkmalMarhaliEncoding = face_recognition.face_encodings(AkmalMarhali)[0]
```

```python
known_faces = [NajwaShihabEncoding,MaruararEncoding,AkmalMarhaliEncoding]
```

```python
face_location = []
face_encodings = []
face_names = []
frame_number = 0
```

```python
while True:
    ret, frame = input_video.read()
    frame_number += 1
    if not ret:
        break
    rgb_frame = frame[:, :, ::-1]
    face_locations = face_recognition.face_locations(rgb_frame)
    face_encodings = face_recognition.face_encodings(rgb_frame, face_locations)
    
    face_names = []
    for face_encoding in face_encodings:
        match = face_recognition.compare_faces(known_faces, face_encoding, tolerance = 0.50)
        
        name = None
        if match[0]:
            name = "Najwa"
        elif match[1]:
            name = "Maruarar"
        elif match[2]:
            name = "Akmal"

    face_names.append(name)

    for (top, right, bottom, left), name in zip(face_locations, face_names):
        if not name:
            continue

#Draw box around the face

        cv2.rectangle(frame, (left,top), (right,bottom), (0,0,255), 2)

#Draw A Label
        cv2.rectangle(frame, (left,bottom - 25), (right, bottom), (0,0,255), cv2.FILLED)
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(frame, name, (left + 6, bottom - 6), font, 0.5, (255,255,255), 1)
    
    print("Writing frame {} / {}".format(frame_number, length))
    output_video.write(frame)
    
input_video.release()
cv2.destroyAllWindows()
```
